```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/homeprices1.csv')
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y = mx + b
# price = m*area + b
135.78767123 * 3300 + 180616.43835616432
```

```python
reg.predict([[5000]])
```

```python
135.78767123 * 5000 + 180616.43835616432
```

```python
%matplotlib inline
plt.xlabel("area (sq ft)")
plt.ylabel("price (US$)")
plt.scatter(df.area, df.price, color="red", marker="+")
```

```python
reg.predict([[5000], [3300]])
```

```python
df_areas = pd.read_csv("/content/areas.csv")
df_areas.head(2)
```

```python
p = reg.predict(df_areas[["area"]])
```

```python
df_areas["price_p"] = p
```

```python
df_areas.head()
```

```python
df_areas.to_csv("prediction.csv", index=False)
```

```python
%matplotlib inline
plt.xlabel("area (sq ft)")
plt.ylabel("price (US$)")
plt.scatter(df.area, df.price, color="red", marker="+")
plt.scatter(df_areas.area, df_areas.price_p, color="black", marker="+")
```
